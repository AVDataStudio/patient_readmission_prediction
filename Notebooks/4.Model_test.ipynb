```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import warnings
warnings.filterwarnings('ignore')

# Load saved model and test data
xgb_model = joblib.load('outputs/xgb_model.pkl')
X_test = pd.read_csv('outputs/X_test.csv')
y_test = pd.read_csv('outputs/y_test.csv')

# Generate predictions
y_pred = xgb_model.predict(X_test)
y_prob = xgb_model.predict_proba(X_test)[:, 1]

print("Model and data loaded")
print(f"Test patients: {len(X_test):,}")
print(f"Predicted readmissions: {y_pred.sum():,}")
print(f"Actual readmissions: {y_test.sum().values[0]:,}")
```
